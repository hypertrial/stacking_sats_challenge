## ü§î Why is this important?
Creating a virtual environment ensures you‚Äôre working with **the exact same setup** (Python version and libraries) that we‚Äôll use to evaluate your submission.

‚úÖ This **prevents** the classic "works on my machine but not on yours" issue.  
‚úÖ It **ensures** fairness and consistency across all participants.

---

## üß† What Is a Virtual Environment?
Think of a virtual environment as a separate container on your machine where we can install all the required tools and libraries, completely isolated from the rest of your system. This way, everything stays consistent and you won't have to worry about version mismatches or breaking your main setup.

‚úÖ Different projects can have different dependencies ‚Äî no conflicts.
‚úÖ Your main Python installation stays clean and untouched.
‚úÖ Everyone working on the same project (or challenge) can use the exact same setup.

> In short:
A virtual environment is your project‚Äôs personal sandbox.
verything stays isolated, predictable, and easy to manage!

# üöÄ Let's Get Started

## üí° Recommended: Use Anaconda (or Miniconda)

If you‚Äôre new to Python or working with data science libraries, we **strongly recommend using Anaconda or Miniconda** for the smoothest setup experience.

It simplifies environment management and avoids common installation issues across platforms.

üëâ **Skip ahead to the Anaconda Guide if you're using Anaconda or Miniconda.**

---

## üõ†Ô∏è Step 1: Install Python 3.10

### Option 1: 

1. Go to [python.org/downloads](https://www.python.org/downloads/).
2. **Download Python 3.10** (important: we use **Python 3.10** for the challenge).
3. During installation, make sure you **check**:

    ```
    ‚úÖ Add Python to PATH
    ```

4. Confirm Python is installed by opening a terminal and typing:

    ```bash
    python --version
    # or
    python3 --version
    ```

You should see something like:

```bash
Python 3.10.x
```

### Option 2: 

- You can also install Python 3.10 via Homebrew:

```bash
brew install python@3.10
brew link --force --overwrite python@3.10
```

---

## üõ†Ô∏è Step 2: Install `virtualenv`

`virtualenv` is a simple tool to create isolated environments.

Install it by running:

```bash
pip install virtualenv
# or
pip3 install virtualenv
```

Verify installation:

```bash
virtualenv --version
```

---

## üõ†Ô∏è Step 3: Create Your Virtual Environment

1. Open your terminal.
2. Navigate to where you want your project folder (e.g., Desktop):

    ```bash
    cd Desktop
    ```

3. Create a new folder for the challenge:

    ```bash
    mkdir stacking-sats
    cd stacking-sats
    ```

4. Inside your project folder, create your virtual environment:

    ```bash
    virtualenv stacking-sats-env
    ```

‚úÖ A new folder called `stacking-sats-env` will appear ‚Äî that's your isolated environment!

---

## üõ†Ô∏è Step 4: Activate the Environment

Activate it depending on your system:

- **Windows:**

    ```bash
    .\stacking-sats-env\Scripts\activate
    ```

- **Mac/Linux:**

    ```bash
    source stacking-sats-env/bin/activate
    ```

‚úÖ When activated, your terminal will look like this:

```bash
(stacking-sats-env) your-computer-name %
```

**Reminder:** Always activate your environment before working on your strategy!

---

## üõ†Ô∏è Step 5: Install Required Libraries

Once your environment is activated:

1. Make sure you have the `stacking_sats_requirements.txt` file ready.
   
   If you don't have it yet, you can create one manually by copying the following into a file called `stacking_sats_requirements.txt`:

    ```
    pandas==2.0.0
    numpy<2.0.0
    matplotlib==3.7.0
    coinmetrics-api-client==2024.2.6.16
    pytest==8.0.0
    scipy==1.10.0
    psutil==6.0.0
    bandit==1.7.0
    safety==3.0.0
    pylint==3.0.0
    pytest-cov==4.1.0
    ```

2. Then install all the required libraries by running:

    ```bash
    pip install -r stacking_sats_requirements.txt
    ```

‚úÖ This ensures you have **the exact versions** needed for the challenge!

---

## üõ†Ô∏è Step 6: (Optional but Recommended) Register the Environment as a Jupyter Kernel

If you plan to work in **Jupyter Notebooks**, you'll want your environment available as a selectable kernel.

While your environment is activated:

1. Install the Jupyter kernel package:

    ```bash
    pip install ipykernel
    ```

2. Register your environment as a kernel:

    ```bash
    python -m ipykernel install --user --name=stacking-sats-env --display-name "Stacking Sats Env"
    ```

‚úÖ Now you can easily select the **"Stacking Sats Env"** kernel when opening your Jupyter notebooks!

---

## üõ†Ô∏è Step 7: Start Building!

You're ready to:

- Develop your strategy
- Test it thoroughly
- Submit it with confidence üí™

Each time you work:

1. Open your terminal.
2. Navigate to your project folder.
3. Activate your virtual environment:

    ```bash
    source stacking-sats-env/bin/activate
    # or
    .\stacking-sats-env\Scripts\activate
    ```

4. Open your notebook and **select the "Stacking Sats Env" kernel**.

‚úÖ This ensures your code runs using the correct environment setup.

---

## ‚ùå Step 8 (Optional): Deactivate When Done

When you're finished working:

```bash
deactivate
```

This will return you to your system‚Äôs normal Python environment.

# ‚ö° Troubleshooting Tips

- **Command not found?**  
  ‚ûî Try `python3` instead of `python`, and `pip3` instead of `pip`.

- **Activation error on Mac/Linux?**  
  ‚ûî Run:

    ```bash
    chmod +x stacking-sats-env/bin/activate
    ```

- **Windows Defender warning?**  
  ‚ûî It's safe to allow if you're following these instructions.

- **Kernel not showing up in Jupyter?**  
  ‚ûî Double-check you installed `ipykernel` and ran the `python -m ipykernel install ...` command.

# üêç Using Anaconda Instead? 

If you‚Äôre a **Conda user**, you can absolutely set up your environment for the Stacking Sats Challenge using `conda` instead of `virtualenv`. In fact, it‚Äôs often easier on macOS or when using data science packages like `numpy`, `scipy`, or `matplotlib`.

This guide ensures your environment mirrors the one used for challenge evaluation:
‚úÖ Python 3.10
‚úÖ Exact versions of required libraries
‚úÖ Isolated setup, reproducible on any system

---

### üõ†Ô∏è Step 1: Create a Conda Environment with Python 3.10

Open your terminal and run:

```bash
conda create -n stacking-sats-env python=3.10
```

This creates a new Conda environment named `stacking-sats-env` with Python 3.10 ‚Äî the official version used in the challenge.

> üí° Why 3.10?
> Because some packages like `scipy==1.10.0` or `matplotlib==3.7.0` don‚Äôt have wheels for newer Python versions like 3.12 ‚Äî using 3.10 ensures compatibility and stability.

---

### üõ†Ô∏è Step 2: Activate the Environment

Once it‚Äôs created, activate it:

```bash
conda activate stacking-sats-env
```

You should now see your prompt change to reflect the active environment:

```bash
(stacking-sats-env) %
```

---

### üõ†Ô∏è Step 3: Install Required Libraries

You‚Äôll need a file called `stacking_sats_requirements.txt` in your project folder. It should contain:

```txt
pandas==2.0.0
numpy<2.0.0
matplotlib==3.7.0
coinmetrics-api-client==2024.2.6.16
pytest==8.0.0
scipy==1.10.0
psutil==6.0.0
bandit==1.7.0
safety==3.0.0
pylint==3.0.0
pytest-cov==4.1.0
```

Then, install everything inside your activated conda env using `pip`:

```bash
pip install -r stacking_sats_requirements.txt
```

> ‚ö†Ô∏è **Why `pip install` inside Conda?**
> Because the challenge requires **exact versions** of packages. Using `pip` ensures you match the exact versions listed ‚Äî some of which may not exist in Conda's default or `conda-forge` channels.

---

### üõ†Ô∏è Step 4 (Optional): Register Environment for Jupyter Notebooks

If you plan to work in Jupyter, install the kernel package:

```bash
pip install ipykernel
python -m ipykernel install --user --name=stacking-sats-env --display-name "Stacking Sats Env"
```

Now when you launch a Jupyter notebook, you‚Äôll see "Stacking Sats Env" as an option in the kernel selector.

---

### üõ†Ô∏è Step 5: Deactivate When Done

When you‚Äôre done working:

```bash
conda deactivate
```

Your prompt will return to the base environment or your system shell.

---

### üß† Final Tips

* Always run `conda activate stacking-sats-env` before developing or testing your strategy.
* Make sure to use `pip install -r ...` from *inside* the activated environment.
* If you're ever unsure whether you're in the right environment, run:

  ```bash
  which python
  python --version
  ```

It should show a Python 3.10 binary path inside your Conda env.
